test emotion analysis

```python
text = business_review1.text.iloc[0]
```

```python
from snownlp import SnowNLP
s = SnowNLP(text)
```

```python
s.sentiments
```

```python
def get_sentiment_cn(text):
    s = SnowNLP(text)
    return s.sentiments
```

```python
business_review1["sentiment"] = business_review1.text.apply(get_sentiment_cn)
```

```python
business_review1
```

Check out the average restaurant situation by state

```python
df = business_review1['sentiment'].groupby(business_review1['state']).mean()*100
```
